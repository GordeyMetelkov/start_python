```python
import sympy
x = sympy.Symbol('x')
y = -12*x**4
print(sympy.solve(y))
```

```python
import sympy
import math
x = sympy.Symbol('x')
y = sympy.sin(sympy.cos(x))
print(sympy.solve(y))
dir_y = y.diff(x)
print(sympy.solve(dir_y))
```

```python
import sympy
import math
x = sympy.Symbol('x')
y = 18*x**3+5*x**2 + 10*x - 30
print(sympy.solve(y))
dir_y = y.diff(x)
print(sympy.solve(dir_y))
```

```python
import numpy as np
import matplotlib.pyplot as plt

y = lambda x: -12*x**4*np.sin(np.cos(x)) - 18*x**3+5*x**2 + 10*x - 30
fig = plt.subplots()

x = np.linspace(-10, 10,50)
plt.grid(axis = 'both')
plt.plot(x, y(x))
plt.show()
```
